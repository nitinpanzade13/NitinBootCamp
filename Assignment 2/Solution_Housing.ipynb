```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
df.info()
```

```python
plt.plot(df['price'],df['area'])
plt.ylabel('price')
plt.xlabel('area')

plt.title('Line plot')
plt.show()
```

```python
plt.plot(df['bedrooms'],df['price'])
plt.show()
```

```python
plt.plot(df['bathrooms'],df['price'])
plt.show()
```

```python
plt.plot(df['stories'],df['price'])
plt.show()
```

```python
plt.plot(df['furnishingstatus'],df['price'])
plt.show()
```

```python
plt.bar(df['area'],df['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(df["bedrooms"],df["price"],label="Price")
plt.legend()
plt.ylabel("Price")
plt.xlabel("Bedrooms")
plt.show()
```

```python
plt.scatter(df["price"],df["area"])
plt.xlabel("price")
plt.ylabel("area")
plt.show()
```

```python
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Price vs Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
sns.barplot(x='furnishingstatus', y='price', data=df,hue="furnishingstatus")
```

```python
pivot_table = df.pivot_table(index='bathrooms', columns='stories', values='price', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', linewidths=0.5)

plt.title('Average Price: Bathrooms vs Stories')
plt.xlabel('Number of Stories')
plt.ylabel('Number of Bathrooms')
plt.show()
```

```python
sns.countplot(x='furnishingstatus', hue='mainroad', data=df)
plt.title('Number of Properties by Furnishing Status and Main Road Access')
plt.xlabel('Furnishing Status')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(x="price",data=df,hue="parking",palette="deep")
plt.title('Distribution of Property Prices by Parking Availability')
plt.xlabel('Price')
plt.ylabel('Count')
```
